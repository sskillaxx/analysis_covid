```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('Covid Data.csv')
```

```python
patient_types = df['PATIENT_TYPE'].value_counts()
plt.figure(figsize=(6,6))
patient_types.index = patient_types.index.map({1: 'вернулись', 2: 'остались лечиться'})
plt.pie(patient_types, labels=patient_types.index, autopct='%1.0f%%', colors=['orange','lightblue'])
plt.title("вернувшиеся домой и оставшиеся лечиться")
plt.show()
```

```python
mean_age_sex = df.groupby('SEX')['AGE'].mean()
mean_age_sex.index = mean_age_sex.index.map({1: 'Женщины', 2: 'Мужчины'})
mean_age_sex.plot(kind='bar', color=['lightgreen','salmon'])
plt.title("статистика по возрасту и полу")
plt.xlabel("пол")
plt.ylabel("возраст")
plt.show()
```

```python
women = df[df['SEX'] == 1]

women['AGE'].hist(bins=20, edgecolor='black') # bins - размерность промежутков, которые мы анализируем 
plt.title("распределение age среди женщин")
plt.show()
```

```python
women = df[df['SEX'] == 1]
women_with_covid = women[women['CLASIFFICATION_FINAL'].between(1, 3)]

women_with_covid['AGE'].hist(bins=20, edgecolor='black') # размерность промежутков, которые мы анализируем 
plt.title("женщины какого AGE имеют форму COVID от 1 до 3")
plt.show()
```

```python
men = df[df['SEX'] == 2]
men_with_covid = men[men['CLASIFFICATION_FINAL'].between(1, 3)]

men_with_covid['AGE'].hist(bins=20, edgecolor='black') # размерность промежутков, которые мы анализируем 
plt.title("мужчины какого AGE имеют форму COVID от 1 до 3")
plt.show()
```

```python
with_covid = df[df['CLASIFFICATION_FINAL'].between(1, 3)]
smokers_with_covid = with_covid[with_covid['TOBACCO'] == 1]
non_smokers_with_covid = with_covid[with_covid['TOBACCO'] == 2]

labels = ['курящие больные', 'некурящие больные']
sizes = [len(smokers_with_covid), len(non_smokers_with_covid)]

plt.figure(figsize=(6, 6))
plt.pie(sizes, labels=labels, colors=['orange','lightblue'], autopct='%1.0f%%')
plt.title('распределение курящих и некурящих среди больных COVID-19')
plt.show()
```

```python
with_covid = df[df['CLASIFFICATION_FINAL'].between(1, 3)]
diabeticals_with_covid = with_covid[with_covid['DIABETES'] == 1]
non_diabeticals_with_covid = with_covid[with_covid['DIABETES'] == 2]

labels = ['больные COVID-19 диабетики', 'больные COVID-19 не диабетики']
sizes = [len(diabeticals_with_covid), len(non_diabeticals_with_covid)]

plt.figure(figsize=(6, 6))
plt.pie(sizes, labels=labels, colors=['orange','lightblue'], autopct='%1.0f%%')
plt.title('распределение диабетиков и не диабетиков среди больных COVID-19')
plt.show()
```

```python
with_diabetes = df[df['DIABETES'] == 1]
with_diabetes_no_covid = with_diabetes[with_diabetes['CLASIFFICATION_FINAL'].between(1, 3)]
with_diabetes_covid = with_diabetes[with_diabetes['CLASIFFICATION_FINAL'] >= 4]

labels = ['диабетики больные COVID-19', 'не диабетики больные COVID-19']
sizes = [len(with_diabetes_covid), len(with_diabetes_no_covid)]

plt.figure(figsize=(6, 6))
plt.pie(sizes, labels=labels, colors=['orange','lightblue'], autopct='%1.0f%%')
plt.title('как часто диабетики болеют COVID-19')
plt.show()
```
